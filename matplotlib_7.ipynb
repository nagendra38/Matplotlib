```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('first plot')
plt.show()
```

```python
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label',{'size':20,'color':'green'})
plt.ylabel('y label')
plt.title('first plot')
plt.show()
```

```python
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label',{'size':20,'color':'green'},labelpad=20)
plt.ylabel('y label')
plt.title('first plot')
plt.show()
```

```python
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label',{'size':20,'color':'green'},labelpad=-20)
plt.ylabel('y label')
plt.title('first plot')
plt.show()
```

```python
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label',{'color':'green'},size=40)
plt.ylabel('y label')
plt.title('first plot')
plt.show()
```

```python
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label',{'color':'green'},color='red')
plt.ylabel('y label')
plt.title('first plot')
plt.show()
```

```python
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label')
plt.ylabel('y label',{'color':'red'},labelpad=20,size=40)
plt.title('first plot')
plt.show()
```
