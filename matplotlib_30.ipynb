```python
import matplotlib.pyplot as plt
x=['science','commerce','arts']
h=[200,300,500]
r=['red','green','orange']
plt.bar(x,h,width=-0.5,color=r)
plt.xlabel('courses')
plt.ylabel('student enrolled')
plt.title('student enrolled for diff courses 2020')
plt.show()
```

```python
import matplotlib.pyplot as plt
y=['science','commerce','arts']
w=[200,300,500]
r=['red','green','orange']
plt.barh(y,w,height=0.5,color=r)
plt.xlabel('courses')
plt.ylabel('student enrolled')
plt.title('student enrolled for diff courses 2020')
plt.show()
```
