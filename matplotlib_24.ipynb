```python
import matplotlib.pyplot as plt
import numpy as np

t=[0,30,45,60,90]
x=[i*(np.pi/180) for i in t]

plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t=[0,30,45,60,90] 
x=[i*(np.pi/180) for i in t]

plt.plot(x,np.sin(x),marker='+')
plt.plot(x,np.cos(x),marker='^')
plt.show()
```

```python
 import matplotlib.pyplot as plt
import numpy as np

t=[0,30,45,60,90]
x=[i*(np.pi/180) for i in t]

plt.plot(t,np.sin(x),marker='+')
plt.plot(t,np.cos(x),marker='^')
plt.xticks(t)
plt.show()
```

```python
 import matplotlib.pyplot as plt
import numpy as np

t=[0,30,45,60,90]
x=[i*(np.pi/180) for i in t]

plt.plot(t,np.sin(x),marker='+')
plt.plot(t,np.cos(x),marker='^')
plt.xticks(t)
plt.title('sin & cos')
plt.xlabel('angle')
plt.ylabel('sin & cos values')
plt.show()
```

```python
 import matplotlib.pyplot as plt
import numpy as np

t=[0,30,45,60,90]
x=[i*(np.pi/180) for i in t]

plt.plot(t,np.sin(x),marker='+',label='sin')
plt.plot(t,np.cos(x),marker='^',label='cos')
plt.xticks(t)
plt.title('sin & cos')
plt.xlabel('angle')
plt.ylabel('sin & cos values')
plt.legend()
plt.show()
```

```python
 import matplotlib.pyplot as plt
import numpy as np

t=np.linspace(-np.pi,np.pi)


plt.plot(t,np.sin(t),marker='+',label='sin')
plt.plot(t,np.cos(t),marker='^',label='cos')

plt.title('sin & cos')
plt.xlabel('angle')
plt.ylabel('sin & cos values')
plt.legend()
plt.show()
```

```python
 import matplotlib.pyplot as plt
import numpy as np

t=np.arange(0,4*np.pi,0.5)


plt.plot(t,np.sin(t),marker='+',label='sin')
plt.plot(t,np.cos(t),marker='^',label='cos')

plt.title('sin & cos')
plt.xlabel('angle')
plt.ylabel('sin & cos values')
plt.legend()
plt.show()
```

```python
 import matplotlib.pyplot as plt
import numpy as np

t=np.arange(0,4*np.pi,0.5)


plt.plot(t,np.sin(t),marker='+',label='sin',linestyle='-.')
plt.plot(t,np.cos(t),marker='^',label='cos',linestyle=':')

plt.title('sin & cos')
plt.xlabel('angle')
plt.ylabel('sin & cos values')
plt.legend()
plt.show()
```
