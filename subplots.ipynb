```python
import matplotlib.pyplot as plt
y1 = [10,20,30,40,50]
y2 = [15,25,35,45,55]
x = [1,2,3,4,5]
plt.stackplot(x,y1,y2,labels=['y1','y2'],baseline='sym')
plt.legend()
plt.show()
```

```python
plt.subplot(2,2,1)
plt.stackplot(x,y1,y2)

plt.subplot(2,2,2)
plt.stackplot(x,y1,y2,labels=['y1','y2'],baseline='sym')

plt.subplot(2,2,3)
plt.subplot(2,2,4)

plt.show()
```

```python
import matplotlib.pyplot as plt
y1 = [10,20,30,40,50]
y2 = [15,25,35,45,55]
x = [1,2,3,4,5]
plt.stackplot(x,y1,y2,labels=['y1','y2'],baseline='weighted_wiggle')
plt.legend()
plt.show()
```

```python
plt.subplot(2,2,1)
plt.stackplot(x,y1,y2)

plt.subplot(2,2,2)
plt.stackplot(x,y1,y2,labels=['y1','y2'],baseline='sym')

plt.subplot(2,2,3)
plt.stackplot(x,y1,y2,labels=['y1','y2'],baseline='weighted_wiggle')

plt.subplot(2,2,4)

plt.show()
```

```python
import matplotlib.pyplot as plt
y1 = [10,20,30,40,50]
y2 = [15,25,35,45,55]
x = [1,2,3,4,5]
plt.fill_between(x,y1,y2)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.stackplot(x,y1,y2)

plt.subplot(2,2,2)
plt.stackplot(x,y1,y2,labels=['y1','y2'],baseline='sym')

plt.subplot(2,2,3)
plt.stackplot(x,y1,y2,labels=['y1','y2'],baseline='weighted_wiggle')

plt.subplot(2,2,4)
plt.fill_between(x,y1,y2)

plt.show()
```
