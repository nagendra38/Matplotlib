```python
import matplotlib.pyplot as plt
year=[1920,1930,1940,1950,1960,1970,1980,1990,2000]
unemp=[9,4,6,2,9,4,6,3,9]
plt.plot(year,unemp,marker='o')
plt.xlabel('Year')
plt.ylabel('Unemoloyment Rate')
plt.title('Yr vs unemployment')
plt.show()
```

```python
import matplotlib.pyplot as plt
year=[1920,1930,1940,1950,1960,1970,1980,1990,2000]
unemp=[9,4,6,2,9,4,6,3,9]
plt.plot(year,unemp,marker='o',linestyle='-.')
plt.xlabel('Year')
plt.ylabel('Unemoloyment Rate')
plt.title('Yr vs unemployment')
plt.show()
```
