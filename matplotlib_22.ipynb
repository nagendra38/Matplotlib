```python
import matplotlib.pyplot as plt
import numpy as np

t = np.array([1,2,3,4])

line1, = plt.plot(t**2,t,color='red',label='labell1')
line2, = plt.plot(t**3,t,color='green',label='labell2')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('first plot')
plt.grid()
plt.tick_params(length=5)
plt.show() 
```
