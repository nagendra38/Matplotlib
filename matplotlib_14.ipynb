```python
import matplotlib.pyplot as plt
import numpy as np

t = np.array([1,2,3,4])

line1, = plt.plot(t**2,t,color='red')
line2, = plt.plot(t**3,t,color='green')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('first plot')
plt.legend([line2,line1],['cubes','squares'])
plt.show() 
```
