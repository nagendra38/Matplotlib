```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('first plot')
plt.show()
```

```python
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('first plot')
plt.minorticks_on()
plt.grid(b=True,which='minor',color='red')
plt.show()
```

```python
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('first plot')
plt.minorticks_on()
plt.grid(b=True,which='minor',color='red')
plt.grid(b=True,which='major',color='green')  
plt.show()
```
