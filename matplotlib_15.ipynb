```python
import matplotlib.pyplot as plt
import numpy as np

t = np.array([1,2,3,4])

line1, = plt.plot(t**2,t,color='red',label='l1')
line2, = plt.plot(t**3,t,color='green',label='l2')
line3, = plt.plot(t,t**2,color='red',label='l3')
line4, = plt.plot(t,t**3,color='green',label='l4')
plt.xlim(0,4)
plt.ylim(0,4)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('first plot')
plt.grid()
plt.legend()
plt.show() 
```
