```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000,3)

plt.hist(x,30)

plt.xlabel('age')
plt.ylabel('count')
plt.title('population age count')

plt.show()   
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000,3)

plt.hist(x,30,histtype='barstacked')

plt.xlabel('age')
plt.ylabel('count')
plt.title('population age count')

plt.show()   
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000,3)

plt.hist(x,30,stacked=True)

plt.xlabel('age')
plt.ylabel('count')
plt.title('population age count')

plt.show()   
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000)
y = np.sin(x)
z = np.cos(x)

plt.hist(x,bins='auto',ec='k')
plt.hist(y,bins='auto',ec='k')
plt.hist(z,bins='auto',ec='k')

plt.xlabel('age')
plt.ylabel('count')
plt.title('population age count')

plt.show()   
```

```python
 import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000)
y = np.sin(x)
z = np.cos(x)

plt.hist(x,bins='auto',ec='k',alpha=0.5)
plt.hist(y,bins='auto',ec='k',alpha=0.5)
plt.hist(z,bins='auto',ec='k',alpha=0.5)

plt.xlabel('age')
plt.ylabel('count')
plt.title('population age count')

plt.show()   
```
