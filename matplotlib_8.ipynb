```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label',{'size':20})
plt.ylabel('y label',size='xx-small')
plt.title('first plot')
plt.show()
```

```python
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label',color='green',size=20,alpha=0.5)
plt.ylabel('y label',{'color':'blue'},size=20,alpha=0)
plt.title('first plot')
plt.show()
```

```python
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label',color='green',backgroundcolor='red',size=20)
plt.ylabel('y label',{'color':'blue','backgroundcolor':'orange'},size=20)
plt.title('first plot')
plt.show()
```

```python
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label',color='green',style='italic',size=20)
plt.ylabel('y label',{'color':'blue'},size=20)
plt.title('first plot')
plt.show()
```

```python
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label',color='green',weight=200,size=20)
plt.ylabel('y label',{'color':'blue','weight':'black'},size=20)
plt.title('first plot')
plt.show()
```

```python
plt.plot([1,2,3],[10,20,30])
plt.xlabel('x label',color='green',family='fantasy',size=20)
plt.ylabel('y label',{'color':'blue'},size=20)
plt.title('first plot')
plt.show()
```
