```python
import matplotlib.pyplot as plt

x = [1,12,22,21,20,21]
y = [1,11,21,31] 
plt.hist(x,y,ec='red')

plt.xlabel('age')
plt.ylabel('count')
plt.title('population age count')
plt.xticks(y)
plt.show()
```
