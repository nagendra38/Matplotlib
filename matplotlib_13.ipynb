```python
import matplotlib.pyplot as plt
import numpy as np

t = np.array([1,2,3,4])

plt.plot(t**2,t,color='red')
plt.plot(t**3,t,color='green')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('first plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.array([1,2,3,4])

plt.plot(t**2,t,color='red',label='squares')
plt.plot(t**3,t,color='green',label='cubes')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('first plot')
plt.legend()
plt.show()
```
