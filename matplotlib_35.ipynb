```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000)

n,bin1,patches = plt.hist(x,bins=10)

color = ['blue','green','#CCCC00','#00FF00','#FFFF00','blue','green','#CCCC00','#00FF00','#FFFF00']

for i in range(len(n)):
    c = color[i]
    patches[i].set_fc(c)

plt.xlabel('age')
plt.ylabel('count')
plt.title('population age count')

plt.show()   
```
