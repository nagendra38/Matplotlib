```python
import matplotlib.pyplot as plt
import numpy as np

x=['USA','Russia','GBR','China','Germany']

gold = [20,10,6,24,30]
silver = [100,50,30,120,150]
bronze = [200,100,60,240,300]

b_bronze = list(np.add(gold,silver))

plt.bar(x,gold,0.4,label='gold')
plt.bar(x,silver,0.4,bottom=gold,label='silver')
plt.bar(x,bronze,0.4,bottom=b_bronze,label='bronze')

plt.xlabel('countries')
plt.ylabel('medals')
plt.title('countries vs medals')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

w=0.4

x=['CSE','ISE','ECE','EEE','MECH']

boys = [200,40,60,240,300]
girls = [200,70,30,120,15]

plt.bar(x,boys,w,label='boys')
plt.bar(x,girls,w,bottom=boys,label='girls')

plt.xlabel('courses')
plt.ylabel('students')
plt.title('students vs courses')
plt.legend()
plt.show()
```

```python
 import matplotlib.pyplot as plt

w=0.4

x=['CSE','ISE','ECE','EEE','MECH']

boys = [200,40,60,240,300]
girls = [200,70,30,120,15]

b = [1,2,3,4,5]
g = [2,3,4,5,6]

plt.bar(x,boys,w,label='boys',yerr=b)
plt.bar(x,girls,w,bottom=boys,label='girls',yerr=g)

plt.xlabel('courses')
plt.ylabel('students')
plt.title('students vs courses')
plt.legend()
plt.show()
```
